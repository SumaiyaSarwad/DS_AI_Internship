```python
# Fill AnnualIncome with median
df['AnnualIncome'] = df['AnnualIncome'].fillna(df['AnnualIncome'].median())

# Fill Education with mode
df['Education'] = df['Education'].fillna(df['Education'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df = df.drop('CustomerID', axis=1)
```

```python
df.head()
```

```python
df.shape
```

## Data Cleaning Decisions

- The AnnualIncome column had 12 missing values (~4.7% of the dataset). Since it is a numerical feature and contains outliers, I filled the missing values using the median to avoid skewing the data.
- The Education column had 12 missing values. Since it is categorical, I filled missing values using the mode (most frequent category).
- I checked for duplicate rows to ensure data integrity and removed them if any were found.
- The CustomerID column was dropped because it is a unique identifier and does not contribute to analytical insights.
    

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
plt.figure()
plt.hist(df['Age'], bins=15)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

### Age Distribution Insight
Most customers fall between the ages of 30 and 45. The distribution appears fairly balanced with no extreme outliers.

```python
plt.figure()
plt.hist(df['AnnualIncome'], bins=15)
plt.title("Annual Income Distribution")
plt.xlabel("Annual Income")
plt.ylabel("Frequency")
plt.show()
```

### Annual Income Insight
Most customers earn below ₹100,000 annually, while a few customers have extremely high incomes, indicating possible outliers.

```python
plt.figure()
df['PreferredDevice'].value_counts().plot(kind='bar')
plt.title("Preferred Device Distribution")
plt.xlabel("Device")
plt.ylabel("Count")
plt.show()
```

### Device Preference Insight
Tablet appears to be the most preferred device among customers, followed by Laptop and Mobile.

```python
plt.figure()
plt.scatter(df['AnnualIncome'], df['SpendingScore'])
plt.title("Income vs Spending Score")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```

### Income vs Spending Insight
There is no strong linear relationship between annual income and spending score, suggesting that higher income does not always mean higher spending.

```python
plt.figure()
sns.boxplot(x='Gender', y='SpendingScore', data=df)
plt.title("Spending Score by Gender")
plt.show()
```

### Gender vs Spending Insight
The spending score distribution between males and females appears relatively similar, with slight variations in median values.

```python
plt.figure(figsize=(10,8))

correlation_matrix = df.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrix Heatmap")
plt.show()
```

## Correlation Matrix Insights

- Age and YearsEmployed show a very strong positive correlation (0.98), which is expected since older individuals typically have more years of employment.
- AnnualIncome has a moderate negative correlation (-0.38) with SpendingScore, indicating that higher income does not necessarily lead to higher spending behavior.
- Most other relationships are weak, suggesting customer purchasing patterns are influenced by multiple independent factors rather than one dominant variable.

# Executive Summary

After performing complete exploratory data analysis, the following key insights were identified:

1. Strong Demographic Relationship:
   Age and YearsEmployed are highly correlated (0.98), confirming logical consistency in the dataset.

2. Income Does Not Drive Spending:
   AnnualIncome shows a moderate negative correlation with SpendingScore, suggesting that higher earning customers do not necessarily spend more.

3. Complex Customer Behavior:
   Most variables show weak correlations, indicating that customer behavior is multi-dimensional and influenced by several independent factors.

Overall, the dataset reveals that customer engagement and spending patterns cannot be predicted using a single variable, making segmentation and behavioral analysis essential for business strategy.
