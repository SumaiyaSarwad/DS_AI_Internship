Exercise 1 — Linear Regression (Single Variable)

Predict Canada's per capita income in year 2020.

Using this dataset: canada_per_capita_income.csv

Build a Linear Regression model.

Use it to predict the per capita income of Canadian citizens.

First predict the value for year 2020.

Then predict the per capita income from 2020 to 2030.

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"B:\AIML Internship\src\Day 26\canada_per_capita_income.csv")
df.head()
```

```python
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(df.year, df['per capita income (US$)'], color='red')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
reg.predict([[2020]])
```

```python
for year in range(2020, 2031):
    print(year, reg.predict([[year]])[0])
```
