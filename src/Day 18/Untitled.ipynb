```python
import sqlite3
import pandas as pd

# Connect to correct database
conn = sqlite3.connect("task2_data_connector.db")

# JOIN query
query = """
SELECT interns.name, interns.track, mentors.mentor_name
FROM interns
INNER JOIN mentors
ON interns.track = mentors.track;
"""

# Load into DataFrame
df = pd.read_sql_query(query, conn)

# Show result
df
```
