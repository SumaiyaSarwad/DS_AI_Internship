The Three Branches of Machine Learning:

1. Supervised Learning:
Supervised Learning is a type of machine learning where the data contains labels or correct answers. The model learns by comparing its predictions with the actual output and improving over time. It is mainly used for classification and regression problems.

2. Unsupervised Learning:
Unsupervised Learning is used when the data does not have labels. The model tries to find hidden patterns, structures, or clusters within the data without being told the correct answers beforehand.

3. Reinforcement Learning:
Reinforcement Learning is a type of learning where an agent interacts with an environment and learns by receiving rewards or penalties based on its actions. The goal is to maximize the total reward over time.
